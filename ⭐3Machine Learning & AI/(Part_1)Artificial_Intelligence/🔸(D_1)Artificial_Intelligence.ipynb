```python
import tkinter as tk
from tkinter import ttk, messagebox
import random
import json
import os
from datetime import datetime
from typing import Optional, List, Tuple, Dict

class TicTacToeGame:
    """Core game logic for Tic-Tac-Toe"""
    
    def __init__(self):
        self.board = [['' for _ in range(3)] for _ in range(3)]
        self.current_player = 'X'
        self.game_over = False
        self.winner = None
        self.moves_history = []
        self.player_x_score = 0
        self.player_o_score = 0
        self.ties = 0
        
    def reset_board(self):
        """Reset the game board"""
        self.board = [['' for _ in range(3)] for _ in range(3)]
        self.current_player = 'X'
        self.game_over = False
        self.winner = None
        self.moves_history = []
        
    def make_move(self, row: int, col: int) -> bool:
        """Make a move on the board"""
        if self.game_over or self.board[row][col] != '':
            return False
            
        self.board[row][col] = self.current_player
        self.moves_history.append((row, col, self.current_player))
        
        if self.check_winner():
            self.game_over = True
            self.winner = self.current_player
            if self.current_player == 'X':
                self.player_x_score += 1
            else:
                self.player_o_score += 1
        elif self.is_board_full():
            self.game_over = True
            self.winner = 'Tie'
            self.ties += 1
        else:
            self.current_player = 'O' if self.current_player == 'X' else 'X'
            
        return True
    
    def check_winner(self) -> bool:
        """Check if there's a winner"""
        # Check rows
        for row in self.board:
            if row[0] == row[1] == row[2] != '':
                return True
        
        # Check columns
        for col in range(3):
            if self.board[0][col] == self.board[1][col] == self.board[2][col] != '':
                return True
        
        # Check diagonals
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != '':
            return True
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != '':
            return True
            
        return False
    
    def is_board_full(self) -> bool:
        """Check if the board is full"""
        for row in self.board:
            for cell in row:
                if cell == '':
                    return False
        return True
    
    def get_available_moves(self) -> List[Tuple[int, int]]:
        """Get all available moves"""
        moves = []
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == '':
                    moves.append((i, j))
        return moves
    
    def get_board_state(self) -> List[List[str]]:
        """Get current board state"""
        return [row[:] for row in self.board]


class AIOpponent:
    """AI opponent with different difficulty levels"""
    
    def __init__(self, difficulty: str = 'medium'):
        self.difficulty = difficulty
        
    def get_move(self, game: TicTacToeGame) -> Optional[Tuple[int, int]]:
        """Get AI move based on difficulty level"""
        available_moves = game.get_available_moves()
        
        if not available_moves:
            return None
            
        if self.difficulty == 'easy':
            return self._easy_move(available_moves)
        elif self.difficulty == 'medium':
            return self._medium_move(game, available_moves)
        else:  # hard
            return self._hard_move(game, available_moves)
    
    def _easy_move(self, moves: List[Tuple[int, int]]) -> Tuple[int, int]:
        """Random move for easy difficulty"""
        return random.choice(moves)
    
    def _medium_move(self, game: TicTacToeGame, moves: List[Tuple[int, int]]) -> Tuple[int, int]:
        """Medium difficulty: sometimes strategic, sometimes random"""
        # Try to win if possible
        for move in moves:
            temp_game = self._simulate_move(game, move, 'O')
            if temp_game.check_winner():
                return move
        
        # Block opponent from winning
        for move in moves:
            temp_game = self._simulate_move(game, move, 'X')
            if temp_game.check_winner():
                return move
        
        # 70% chance to make a strategic move, 30% random
        if random.random() < 0.7:
            # Prefer center and corners
            strategic_moves = []
            for move in moves:
                if move in [(1, 1), (0, 0), (0, 2), (2, 0), (2, 2)]:
                    strategic_moves.append(move)
            if strategic_moves:
                return random.choice(strategic_moves)
        
        return random.choice(moves)
    
    def _hard_move(self, game: TicTacToeGame, moves: List[Tuple[int, int]]) -> Tuple[int, int]:
        """Hard difficulty using minimax algorithm"""
        best_score = -float('inf')
        best_move = None
        
        for move in moves:
            temp_game = self._simulate_move(game, move, 'O')
            score = self._minimax(temp_game, 0, False)
            if score > best_score:
                best_score = score
                best_move = move
        
        return best_move if best_move else random.choice(moves)
    
    def _minimax(self, game: TicTacToeGame, depth: int, is_maximizing: bool) -> int:
        """Minimax algorithm implementation"""
        if game.check_winner():
            return 10 - depth if game.winner == 'O' else depth - 10
        elif game.is_board_full():
            return 0
        
        if is_maximizing:
            best_score = -float('inf')
            for move in game.get_available_moves():
                temp_game = self._simulate_move(game, move, 'O')
                score = self._minimax(temp_game, depth + 1, False)
                best_score = max(best_score, score)
            return best_score
        else:
            best_score = float('inf')
            for move in game.get_available_moves():
                temp_game = self._simulate_move(game, move, 'X')
                score = self._minimax(temp_game, depth + 1, True)
                best_score = min(best_score, score)
            return best_score
    
    def _simulate_move(self, game: TicTacToeGame, move: Tuple[int, int], player: str) -> TicTacToeGame:
        """Create a copy of the game with a simulated move"""
        temp_game = TicTacToeGame()
        temp_game.board = [row[:] for row in game.board]
        temp_game.current_player = player
        temp_game.make_move(move[0], move[1])
        return temp_game


class GameHistory:
    """Manage game history and statistics"""
    
    def __init__(self, filename: str = 'tictactoe_history.json'):
        self.filename = filename
        self.history = self._load_history()
        
    def _load_history(self) -> List[Dict]:
        """Load game history from file"""
        if os.path.exists(self.filename):
            try:
                with open(self.filename, 'r') as f:
                    return json.load(f)
            except:
                return []
        return []
    
    def save_game(self, game: TicTacToeGame, mode: str, difficulty: str = None):
        """Save a completed game to history"""
        game_data = {
            'timestamp': datetime.now().isoformat(),
            'winner': game.winner,
            'moves': len(game.moves_history),
            'player_x_score': game.player_x_score,
            'player_o_score': game.player_o_score,
            'ties': game.ties,
            'mode': mode,
            'difficulty': difficulty
        }
        
        self.history.append(game_data)
        self._save_history()
    
    def _save_history(self):
        """Save history to file"""
        with open(self.filename, 'w') as f:
            json.dump(self.history, f, indent=2)
    
    def get_statistics(self) -> Dict:
        """Get overall game statistics"""
        if not self.history:
            return {'total_games': 0, 'x_wins': 0, 'o_wins': 0, 'ties': 0}
        
        stats = {
            'total_games': len(self.history),
            'x_wins': 0,
            'o_wins': 0,
            'ties': 0
        }
        
        for game in self.history:
            if game['winner'] == 'X':
                stats['x_wins'] += 1
            elif game['winner'] == 'O':
                stats['o_wins'] += 1
            else:
                stats['ties'] += 1
        
        return stats


class TicTacToeGUI:
    """Main GUI application for Tic-Tac-Toe"""
    
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Tic-Tac-Toe Pro")
        self.root.geometry("800x600")
        self.root.configure(bg='#2c3e50')
        
        self.game = TicTacToeGame()
        self.ai = None
        self.game_history = GameHistory()
        self.game_mode = 'pvp'  # pvp, pve, eve
        self.ai_difficulty = 'medium'
        
        self.setup_styles()
        self.create_widgets()
        self.update_display()
        
    def setup_styles(self):
        """Setup custom styles for widgets"""
        self.style = ttk.Style()
        self.style.theme_use('clam')
        
        # Configure colors
        self.colors = {
            'bg': '#2c3e50',
            'fg': '#ecf0f1',
            'button_bg': '#3498db',
            'button_fg': '#ffffff',
            'board_bg': '#34495e',
            'x_color': '#e74c3c',
            'o_color': '#3498db',
            'highlight': '#1abc9c'
        }
        
    def create_widgets(self):
        """Create all GUI widgets"""
        # Main container
        main_frame = ttk.Frame(self.root, padding="20")
        main_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))
        
        # Title
        title_label = tk.Label(
            main_frame,
            text="TIC-TAC-TOE PRO",
            font=('Arial', 28, 'bold'),
            fg=self.colors['highlight'],
            bg=self.colors['bg']
        )
        title_label.grid(row=0, column=0, columnspan=3, pady=(0, 20))
        
        # Left panel - Game board
        board_frame = ttk.LabelFrame(main_frame, text="Game Board", padding="10")
        board_frame.grid(row=1, column=0, rowspan=2, padx=(0, 20), sticky=(tk.W, tk.E, tk.N, tk.S))
        
        self.board_buttons = []
        for i in range(3):
            row_buttons = []
            for j in range(3):
                btn = tk.Button(
                    board_frame,
                    text='',
                    font=('Arial', 40, 'bold'),
                    width=3,
                    height=1,
                    bg=self.colors['board_bg'],
                    fg=self.colors['fg'],
                    activebackground='#4a6fa5',
                    command=lambda r=i, c=j: self.on_board_click(r, c)
                )
                btn.grid(row=i, column=j, padx=5, pady=5, sticky=(tk.W, tk.E, tk.N, tk.S))
                row_buttons.append(btn)
            self.board_buttons.append(row_buttons)
        
        # Center panel - Controls
        control_frame = ttk.LabelFrame(main_frame, text="Game Controls", padding="10")
        control_frame.grid(row=1, column=1, padx=10, pady=(0, 10), sticky=(tk.W, tk.E, tk.N, tk.S))
        
        # Game mode selection
        ttk.Label(control_frame, text="Game Mode:").grid(row=0, column=0, sticky=tk.W, pady=5)
        self.mode_var = tk.StringVar(value='pvp')
        modes = [('Player vs Player', 'pvp'), ('Player vs AI', 'pve'), ('AI vs AI', 'eve')]
        for i, (text, value) in enumerate(modes):
            ttk.Radiobutton(
                control_frame,
                text=text,
                variable=self.mode_var,
                value=value,
                command=self.on_mode_change
            ).grid(row=i+1, column=0, sticky=tk.W, pady=2)
        
        # AI difficulty
        ttk.Label(control_frame, text="AI Difficulty:").grid(row=4, column=0, sticky=tk.W, pady=(10, 5))
        self.difficulty_var = tk.StringVar(value='medium')
        difficulties = [('Easy', 'easy'), ('Medium', 'medium'), ('Hard', 'hard')]
        for i, (text, value) in enumerate(difficulties):
            ttk.Radiobutton(
                control_frame,
                text=text,
                variable=self.difficulty_var,
                value=value,
                command=self.on_difficulty_change
            ).grid(row=i+5, column=0, sticky=tk.W, pady=2)
        
        # Control buttons
        button_frame = ttk.Frame(control_frame)
        button_frame.grid(row=8, column=0, pady=(20, 0))
        
        ttk.Button(
            button_frame,
            text="New Game",
            command=self.new_game
        ).pack(side=tk.LEFT, padx=5)
        
        ttk.Button(
            button_frame,
            text="Reset Scores",
            command=self.reset_scores
        ).pack(side=tk.LEFT, padx=5)
        
        # Right panel - Game info
        info_frame = ttk.LabelFrame(main_frame, text="Game Information", padding="10")
        info_frame.grid(row=1, column=2, padx=(20, 0), sticky=(tk.W, tk.E, tk.N, tk.S))
        
        # Score display
        self.score_label = tk.Label(
            info_frame,
            text="Scores:\nX: 0\nO: 0\nTies: 0",
            font=('Arial', 14),
            justify=tk.LEFT,
            bg=self.colors['bg'],
            fg=self.colors['fg']
        )
        self.score_label.pack(pady=(0, 20))
        
        # Current player display
        self.player_label = tk.Label(
            info_frame,
            text="Current Player: X",
            font=('Arial', 16, 'bold'),
            bg=self.colors['bg'],
            fg=self.colors['highlight']
        )
        self.player_label.pack(pady=(0, 20))
        
        # Status display
        self.status_label = tk.Label(
            info_frame,
            text="Game in progress",
            font=('Arial', 12),
            bg=self.colors['bg'],
            fg=self.colors['fg']
        )
        self.status_label.pack(pady=(0, 20))
        
        # Statistics button
        ttk.Button(
            info_frame,
            text="View Statistics",
            command=self.show_statistics
        ).pack(pady=10)
        
        # Move history
        history_frame = ttk.LabelFrame(main_frame, text="Move History", padding="10")
        history_frame.grid(row=2, column=1, columnspan=2, pady=(10, 0), sticky=(tk.W, tk.E, tk.N, tk.S))
        
        self.history_text = tk.Text(
            history_frame,
            height=5,
            width=40,
            font=('Courier', 10),
            bg='#1a1a1a',
            fg='#ffffff'
        )
        self.history_text.pack()
        
        # Configure grid weights
        self.root.columnconfigure(0, weight=1)
        self.root.rowconfigure(0, weight=1)
        main_frame.columnconfigure(0, weight=1)
        main_frame.columnconfigure(1, weight=0)
        main_frame.columnconfigure(2, weight=0)
        main_frame.rowconfigure(1, weight=1)
        
    def on_board_click(self, row: int, col: int):
        """Handle board button clicks"""
        if self.game.game_over:
            return
            
        if self.game_mode == 'eve':
            messagebox.showinfo("AI vs AI", "AI vs AI mode is running automatically!")
            return
            
        if self.game.make_move(row, col):
            self.update_display()
            
            # Check if game is over
            if self.game.game_over:
                self.end_game()
            # AI move in PvE mode
            elif self.game_mode == 'pve' and self.game.current_player == 'O':
                self.root.after(500, self.make_ai_move)
    
    def make_ai_move(self):
        """Make AI move in PvE mode"""
        if not self.game.game_over and self.game.current_player == 'O':
            if self.ai is None:
                self.ai = AIOpponent(self.ai_difficulty)
            
            move = self.ai.get_move(self.game)
            if move:
                self.game.make_move(move[0], move[1])
                self.update_display()
                
                if self.game.game_over:
                    self.end_game()
    
    def on_mode_change(self):
        """Handle game mode change"""
        self.game_mode = self.mode_var.get()
        if self.game_mode == 'eve':
            self.start_ai_vs_ai()
        else:
            self.new_game()
    
    def on_difficulty_change(self):
        """Handle difficulty change"""
        self.ai_difficulty = self.difficulty_var.get()
        if self.ai:
            self.ai.difficulty = self.ai_difficulty
    
    def start_ai_vs_ai(self):
        """Start AI vs AI simulation"""
        self.new_game()
        self.ai = AIOpponent(self.ai_difficulty)
        self.simulate_ai_vs_ai()
    
    def simulate_ai_vs_ai(self):
        """Simulate AI vs AI game"""
        if self.game_mode != 'eve' or self.game.game_over:
            return
            
        # AI X move
        if self.game.current_player == 'X':
            move = self.ai.get_move(self.game)
            if move:
                self.game.make_move(move[0], move[1])
                self.update_display()
        
        # AI O move
        if not self.game.game_over and self.game.current_player == 'O':
            move = self.ai.get_move(self.game)
            if move:
                self.game.make_move(move[0], move[1])
                self.update_display()
        
        # Continue simulation if game not over
        if not self.game.game_over:
            self.root.after(1000, self.simulate_ai_vs_ai)
        else:
            self.end_game()
    
    def update_display(self):
        """Update all display elements"""
        # Update board buttons
        for i in range(3):
            for j in range(3):
                cell_value = self.game.board[i][j]
                btn = self.board_buttons[i][j]
                btn.config(text=cell_value)
                
                # Color coding
                if cell_value == 'X':
                    btn.config(fg=self.colors['x_color'])
                elif cell_value == 'O':
                    btn.config(fg=self.colors['o_color'])
                else:
                    btn.config(fg=self.colors['fg'])
        
        # Update scores
        self.score_label.config(
            text=f"Scores:\nX: {self.game.player_x_score}\nO: {self.game.player_o_score}\nTies: {self.game.ties}"
        )
        
        # Update current player
        player_text = "Current Player: "
        if self.game.game_over:
            player_text = "Game Over - "
        player_text += self.game.current_player
        
        self.player_label.config(text=player_text)
        
        # Update status
        if self.game.game_over:
            if self.game.winner == 'Tie':
                self.status_label.config(text="It's a Tie!", fg='#f39c12')
            else:
                self.status_label.config(text=f"Player {self.game.winner} Wins!", fg=self.colors['highlight'])
        else:
            self.status_label.config(text="Game in progress", fg=self.colors['fg'])
        
        # Update move history
        self.history_text.delete(1.0, tk.END)
        for idx, (row, col, player) in enumerate(self.game.moves_history, 1):
            self.history_text.insert(tk.END, f"Move {idx}: {player} at ({row}, {col})\n")
    
    def new_game(self):
        """Start a new game"""
        self.game.reset_board()
        self.update_display()
        
        # Auto-start AI vs AI
        if self.game_mode == 'eve':
            self.root.after(1000, self.simulate_ai_vs_ai)
    
    def reset_scores(self):
        """Reset all scores"""
        if messagebox.askyesno("Reset Scores", "Are you sure you want to reset all scores?"):
            self.game.player_x_score = 0
            self.game.player_o_score = 0
            self.game.ties = 0
            self.update_display()
    
    def end_game(self):
        """Handle game end"""
        # Save to history
        difficulty = self.ai_difficulty if self.game_mode != 'pvp' else None
        self.game_history.save_game(self.game, self.game_mode, difficulty)
        
        # Show message
        if self.game.winner == 'Tie':
            message = "Game ended in a tie!"
        else:
            message = f"Player {self.game.winner} wins!"
        
        if self.game_mode == 'eve':
            message = f"AI vs AI complete!\n{message}"
        
        messagebox.showinfo("Game Over", message)
    
    def show_statistics(self):
        """Display game statistics"""
        stats = self.game_history.get_statistics()
        
        stats_window = tk.Toplevel(self.root)
        stats_window.title("Game Statistics")
        stats_window.geometry("400x300")
        stats_window.configure(bg=self.colors['bg'])
        
        # Statistics display
        stats_text = f"""
        Overall Statistics:
        --------------------
        Total Games: {stats['total_games']}
        Player X Wins: {stats['x_wins']}
        Player O Wins: {stats['o_wins']}
        Ties: {stats['ties']}
        
        Current Session:
        --------------------
        X Wins: {self.game.player_x_score}
        O Wins: {self.game.player_o_score}
        Ties: {self.game.ties}
        """
        
        stats_label = tk.Label(
            stats_window,
            text=stats_text,
            font=('Courier', 12),
            justify=tk.LEFT,
            bg=self.colors['bg'],
            fg=self.colors['fg'],
            padx=20,
            pady=20
        )
        stats_label.pack()
        
        # Close button
        ttk.Button(
            stats_window,
            text="Close",
            command=stats_window.destroy
        ).pack(pady=10)
    
    def run(self):
        """Run the application"""
        self.root.mainloop()


def main():
    """Main entry point"""
    print("Starting Tic-Tac-Toe Pro...")
    print("Features:")
    print("- Player vs Player mode")
    print("- Player vs AI (Easy/Medium/Hard)")
    print("- AI vs AI simulation")
    print("- Game history and statistics")
    print("- Beautiful GUI interface")
    print("- Score tracking")
    
    app = TicTacToeGUI()
    app.run()


if __name__ == "__main__":
    main()
```
