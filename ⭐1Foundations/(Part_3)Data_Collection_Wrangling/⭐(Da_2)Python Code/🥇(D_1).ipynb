# ⭐: Missing values check & Cleaning

```python
import numpy as np
new_df = df.replace("NaN", np.nan)
new_df.info()
```

```python
df.isnull().sum()
```

```python
print(df[["director", "cast", "country", "date_added", "rating"]].isnull().sum())
print(df[["director", "cast", "country", "date_added", "rating"]].shape)
```

```python
df["director"].fillna("Unknown", inplace=True)
df["cast"].fillna("Not Specified", inplace=True)
df["country"].fillna("Unknown", inplace=True)
df["date_added"].fillna("Unknown", inplace=True)
df["rating"].fillna("Not Rated", inplace=True)
```

```python
# Categorical columns (text/object type)
categorical_cols = df.select_dtypes(include='object').columns
print("Categorical columns:\n", categorical_cols)

# Numerical columns (int/float type)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns:\n", numerical_cols)
```

```python
print(df[numerical_cols].mean(axis=1, skipna=False))
```

```python
# Example: Most common rating
print(df['rating'].value_counts(5))

# Most frequent director
print(df['director'].value_counts().head(5))
```

# ⭐: Map / Encode Categorical Columns

```python
# Map "Movie" → 0, "TV Show" → 1
df['type_mapped'] = df['type'].map({'Movie': 0,
                                    'TV Show': 1})

df[['type', 'type_mapped']].head()
```

# ⭐: Import sqlite3

```python
import sqlite3
import pandas as pd

# Create an in-memory database
conn = sqlite3.connect(':memory:')

# Create a sample table
conn.execute("""
CREATE TABLE movies (
    id INTEGER,
    title TEXT,
    director TEXT,
    year INTEGER
)
""")

# Insert sample data
conn.executemany("INSERT INTO movies VALUES (?, ?, ?, ?)", [
    (1, "Inception", "Christopher Nolan", 2010),
    (2, "Titanic", "James Cameron", 1997),
    (3, "The Matrix", "Lana Wachowski", 1999)
])

# Query the table into a DataFrame
data = pd.read_sql_query("SELECT * FROM movies", conn)
print(data)
```

```python
df = data
print(df)
```
