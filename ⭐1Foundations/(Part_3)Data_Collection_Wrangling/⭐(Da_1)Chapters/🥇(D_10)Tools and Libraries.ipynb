```python

# 3Ô∏è‚É£ Quick Python Example ‚Äì Using Pandas for Wrangling
import pandas as pd
import numpy as np

# Raw Data
data = {
    "Name": ["Alice", "Bob", None, "Emma", "John", "Alice"],
    "Age": [25, 30, 22, None, 28, 25],
    "City": ["Paris", "London", "New York", "paris", None, "Paris"]
}
df = pd.DataFrame(data)
print("\nRaw Data:")
print(df)

# Wrangling with Pandas
df["Name"].fillna("Unknown", inplace=True)   # Handle missing
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["City"].fillna("Unknown", inplace=True)
df.drop_duplicates(inplace=True)
df["City"] = df["City"].str.title()         # Standardize text

print("\nCleaned Data with Pandas:")
print(df)
```
