```python

# 1Ô∏è‚É£ Definition
definition = """
Primary Data Collection refers to gathering data directly from original sources 
for a specific research purpose. It is firsthand, accurate, and tailored.
"""
print(definition)
```

```python

# 2Ô∏è‚É£ Techniques Overview
techniques = {
    "Surveys / Questionnaires": "Structured set of questions to collect responses from a target group.",
    "Interviews": "Direct conversation (face-to-face, phone, or virtual) to gather in-depth insights.",
    "Observations": "Monitoring behavior, events, or processes in natural or controlled settings.",
    "Experiments": "Conducting controlled tests to measure effects of variables.",
    "Focus Groups": "Group discussions guided by a moderator to understand opinions and attitudes."
}

print("Primary Data Collection Techniques:")
for method, desc in techniques.items():
    print(f"- {method}: {desc}")
```

```python

# 3Ô∏è‚É£ Quick Examples
# a) Survey Example
survey_responses = ["Yes", "No", "Yes", "Yes"]  # Responses collected from people

# b) Observation Example
observed_behavior = ["Customer browses", "Customer buys", "Customer leaves"]

# c) Experiment Example
# Suppose testing fertilizer on plant growth
control_group = [10, 12, 11]  # No fertilizer
test_group = [15, 18, 17]     # Fertilizer applied

print("\nSurvey Responses:", survey_responses)
print("Observed Behavior:", observed_behavior)
print("Experiment Results - Control:", control_group, "Test:", test_group)
```
