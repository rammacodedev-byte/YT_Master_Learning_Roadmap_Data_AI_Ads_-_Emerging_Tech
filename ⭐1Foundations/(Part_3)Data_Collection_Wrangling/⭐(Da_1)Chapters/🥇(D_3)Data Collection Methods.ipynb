# üóÇÔ∏è Data Collection Methods ‚Äì Primary & Secondary üì•


```python

# 1Ô∏è‚É£ What is Data Collection?
# Collecting data from various sources to analyze, model, or make decisions.

definition = """
Data Collection Methods are techniques used to gather information for analysis.
They are broadly classified into Primary and Secondary methods.
"""
print(definition)
```

```python

# 2Ô∏è‚É£ Primary Data Collection üõ†Ô∏è
# Data collected firsthand by the researcher
primary_methods = {
    "Surveys / Questionnaires": "Structured questions to gather information from people",
    "Interviews": "One-on-one or group discussions to collect detailed insights",
    "Observations": "Watching behavior, events, or processes directly",
    "Experiments": "Controlled tests to measure effects of variables"
}

print("Primary Data Collection Methods:")
for method, desc in primary_methods.items():
    print(f"- {method}: {desc}")
```

```python

# 3Ô∏è‚É£ Secondary Data Collection üìö
# Data collected from existing sources, not gathered directly
secondary_methods = {
    "Published Reports": "Books, research papers, articles",
    "Databases": "Government data, company records, Kaggle datasets",
    "Websites / APIs": "Scraped or downloaded data from the internet",
    "Social Media": "Data from Twitter, Instagram, LinkedIn, etc."
}

print("\nSecondary Data Collection Methods:")
for method, desc in secondary_methods.items():
    print(f"- {method}: {desc}")
```

```python

# 4Ô∏è‚É£ Quick Tips üîπ
# - Primary data is more accurate and specific but time-consuming
# - Secondary data is faster and cheaper but may be outdated or incomplete
# - Always check reliability and validity of data sources
# - Ethical considerations: privacy & consent
```
