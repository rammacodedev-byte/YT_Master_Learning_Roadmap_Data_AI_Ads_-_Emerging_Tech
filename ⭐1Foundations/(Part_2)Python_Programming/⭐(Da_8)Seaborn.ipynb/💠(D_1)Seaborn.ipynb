```python

# üîπ Quick Tips:
# - Seaborn gives better default styling
# - Works directly with Pandas DataFrames
# - hue ‚Üí compare categories easily
# - Best for EDA & statistical insights
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load built-in dataset üìä
df = sns.load_dataset("tips")

# 1Ô∏è‚É£ Scatter Plot üîµ
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Scatter Plot")
plt.show()
```

```python

# 2Ô∏è‚É£ Line Plot üìà
sns.lineplot(data=df, x="size", y="total_bill")
plt.title("Line Plot")
plt.show()
```

```python

# 3Ô∏è‚É£ Bar Plot üìä
sns.barplot(data=df, x="day", y="total_bill")
plt.title("Bar Plot")
plt.show()
```

```python

# 4Ô∏è‚É£ Box Plot üì¶
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Box Plot")
plt.show()
```

```python

# 5Ô∏è‚É£ Histogram üìâ
sns.histplot(data=df, x="total_bill", bins=20)
plt.title("Histogram")
plt.show()
```

```python

# 6Ô∏è‚É£ Heatmap üî•
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
