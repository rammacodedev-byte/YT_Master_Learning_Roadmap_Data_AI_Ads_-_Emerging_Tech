```python

# üîπ Quick Tips:
# - NumPy arrays are faster than Python lists
# - Use vectorized operations (no loops)
# - reshape() changes dimensions
# - Use NumPy for math & matrix operations
```

```python

import numpy as np

# 1Ô∏è‚É£ Create NumPy Arrays üì¶
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([[1, 2], [3, 4]])
print(arr1)
print(arr2)
```

```python

# 2Ô∏è‚É£ Array Properties üîç
print(arr1.ndim)     # Number of dimensions
print(arr1.shape)    # Shape of array
print(arr1.size)     # Total elements
print(arr1.dtype)    # Data type
```

```python

# 3Ô∏è‚É£ Special Arrays ‚öôÔ∏è
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))
identity = np.eye(3)
print(zeros)
print(ones)
print(identity)
```

```python

# 4Ô∏è‚É£ Range & Random Arrays üéØ
nums = np.arange(1, 10, 2)
rand = np.random.rand(3, 3)
print(nums)
print(rand)
```

```python

# 5Ô∏è‚É£ Array Operations ‚ûï‚ûñ‚úñÔ∏è‚ûó
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
print(a ** 2)
```

```python

# 6Ô∏è‚É£ Indexing & Slicing ‚úÇÔ∏è
arr = np.array([10, 20, 30, 40, 50])
print(arr[1])
print(arr[1:4])
```

```python

# 7Ô∏è‚É£ Reshape & Flatten üîÑ
matrix = np.arange(1, 7).reshape(2, 3)
print(matrix)
print(matrix.flatten())
```

```python

# 8Ô∏è‚É£ Statistical Functions üìä
data = np.array([10, 20, 30, 40, 50])
print(np.mean(data))
print(np.sum(data))
print(np.min(data))
print(np.max(data))
print(np.std(data))
```
