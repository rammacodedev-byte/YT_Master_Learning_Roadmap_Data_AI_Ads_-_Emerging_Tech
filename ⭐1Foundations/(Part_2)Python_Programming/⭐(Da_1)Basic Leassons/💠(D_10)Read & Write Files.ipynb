# üóÇÔ∏è Python Read & Write Files


```python
# üîπ Quick Tips:
# - Use 'with' to auto-close files
# - Always specify correct file mode
# - Use strip() to remove extra spaces/newlines
# - Handle FileNotFoundError when reading files
```

```python
# üîπ File Modes:
# - 'r' ‚Üí read
# - 'w' ‚Üí write (overwrite)
# - 'a' ‚Üí append
# - 'r+' ‚Üí read & write
```

```python

# 1Ô∏è‚É£ Writing to a File ‚úçÔ∏è
with open("sample.txt", "w") as file:
    file.write("Hello, Python!\n")
    file.write("File handling is easy.")
```

```python

# 2Ô∏è‚É£ Reading from a File üìñ
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
```

```python

# 3Ô∏è‚É£ Reading Line by Line üìÑ
with open("sample.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python

# 4Ô∏è‚É£ Appending to a File ‚ûï
with open("sample.txt", "a") as file:
    file.write("\nThis line is appended.")
```

```python

# 5Ô∏è‚É£ Reading Specific Lines üß©
with open("sample.txt", "r") as file:
    print(file.readline())    # Read first line
    print(file.readlines())   # Read all lines as a list
```

```python

# 6Ô∏è‚É£ Writing Multiple Lines üìö
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open("multi.txt", "w") as file:
    file.writelines(lines)
```
