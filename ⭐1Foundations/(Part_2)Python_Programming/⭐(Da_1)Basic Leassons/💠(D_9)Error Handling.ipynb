# üóÇÔ∏è Python Error Handling (Exception Handling)


```python

# üîπ Quick Tips:
# - try ‚Üí wrap risky code
# - except ‚Üí handle errors
# - else ‚Üí runs if no error
# - finally ‚Üí always executes
# - raise ‚Üí trigger your own exception
# - Custom exceptions ‚Üí create meaningful errors
```

```python

# 1Ô∏è‚É£ Basic try‚Äìexcept üöß
try:
    x = int("10")
    print(x)
except ValueError:
    print("Conversion failed")
```

```python

# 2Ô∏è‚É£ Handling Multiple Exceptions üß©
try:
    a = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
except TypeError:
    print("Type error occurred")
```

```python

# 3Ô∏è‚É£ try‚Äìexcept‚Äìelse üéØ
try:
    num = int("5")
except ValueError:
    print("Invalid number")
else:
    print("Conversion successful:", num)
```

```python

# 4Ô∏è‚É£ try‚Äìexcept‚Äìfinally üîí
try:
    file = open("data.txt", "r")
    print(file.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Execution completed")
```

```python

# 5Ô∏è‚É£ Raising Custom Exceptions üöÄ
age = -5

try:
    if age < 0:
        raise ValueError("Age cannot be negative")
    print("Valid age:", age)

except ValueError as e:
    print("Error:", e)

```

```python

# 6Ô∏è‚É£ Custom Exception Class üß†
class CustomError(Exception):
    pass

try:
    raise CustomError("This is a custom error")
except CustomError as e:
    print("Custom Exception:", e)
```
