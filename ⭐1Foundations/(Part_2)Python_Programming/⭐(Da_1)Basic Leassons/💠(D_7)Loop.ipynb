# üóÇÔ∏è Python Loops

```python

# üîπ Quick Tips:
# - for ‚Üí iterate over sequences
# - while ‚Üí run until condition is false
# - break ‚Üí stop loop immediately
# - continue ‚Üí skip current iteration
# - loop else ‚Üí runs if loop ends normally
# - nested loops ‚Üí loop inside another loop
```

```python


# 1Ô∏è‚É£ for Loop üîÅ
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Using range()
for i in range(1, 6):
    print(i)
```

```python

# 2Ô∏è‚É£ while Loop ‚è≥
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python

# 3Ô∏è‚É£ break Statement üõë
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python

# 4Ô∏è‚É£ continue Statement ‚è≠Ô∏è
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python

# 5Ô∏è‚É£ else with Loop üéØ
for i in range(1, 4):
    print(i)
else:
    print("Loop finished successfully")
```

```python

# 6Ô∏è‚É£ Nested Loop ü™ú
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i={i}, j={j}")

```
