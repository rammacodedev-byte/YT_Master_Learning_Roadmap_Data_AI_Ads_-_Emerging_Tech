```python

# üîπ Quick Tips:
# - Sets store only unique values
# - Sets are unordered (no indexing)
# - Use sets for fast membership checking
# - Use frozenset for immutable sets
# - Best for removing duplicates & set math
```

```python
# 1Ô∏è‚É£ Create a Set
numbers = {1, 2, 3, 4, 4, 5}
print(numbers)              # Duplicates removed automatically

# Empty set (important!)
empty_set = set()           # {} creates an empty dictionary
print(type(empty_set))
```

```python
# 2Ô∏è‚É£ Add & Remove Elements
numbers.add(6)              # Add element
numbers.update([7, 8])      # Add multiple elements
numbers.remove(3)           # Remove (error if not found)
numbers.discard(10)         # Remove (no error if not found)
print(numbers)
```

```python
# 3Ô∏è‚É£ Set Operations üîÑ
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)                # Union
print(A & B)                # Intersection
print(A - B)                # Difference
print(A ^ B)                # Symmetric Difference
```

```python
# 4Ô∏è‚É£ Membership & Length
print(2 in A)               # Membership check
print(len(A))               # Number of elements
```

```python

# 5Ô∏è‚É£ Nested Sets (Use frozenset)
set1 = {frozenset([1, 2]), frozenset([3, 4])}
print(set1)
```
