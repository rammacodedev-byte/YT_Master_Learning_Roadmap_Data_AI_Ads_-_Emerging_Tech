```python
# ğŸ”¹ Quick Tips:
# - Lists â†’ ordered & changeable
# - Tuples â†’ ordered & fixed
# - Sets â†’ unique & unordered
# - Dictionaries â†’ key-value pairs
# - Nesting â†’ combine multiple structures
```

```python
# ğŸ—‚ï¸ Python Data Structures

# 1ï¸âƒ£ List â€“ Ordered & Mutable ğŸ“‹
fruits = ["apple", "banana", "cherry"]
print(fruits)             # ['apple', 'banana', 'cherry']
fruits.append("orange")   # Add element
fruits.remove("banana")   # Remove element
print(fruits[0])          # Access first item
print(fruits[-1])         # Access last item
print(fruits[1:3])        # Slicing
```

```python

# 2ï¸âƒ£ Tuple â€“ Ordered & Immutable ğŸ”’
coordinates = (10, 20)
print(coordinates[0])      # Access item
# coordinates[0] = 5       # âŒ Error: tuples cannot be changed
```

```python

# 3ï¸âƒ£ Set â€“ Unordered & Unique ğŸ”¹
colors = {"red", "green", "blue"}
colors.add("yellow")       # Add element
colors.discard("green")    # Remove element
print("red" in colors)     # Membership check
```

```python

# 4ï¸âƒ£ Dictionary â€“ Key-Value Pairs ğŸ—ï¸
person = {"name": "Alice", "age": 25, "city": "Paris"}
print(person["name"])      # Access by key
person["age"] = 26         # Update value
person["job"] = "Engineer" # Add new key-value
print(person.keys())       # All keys
print(person.values())     # All values
```

```python

# 5ï¸âƒ£ Nested Data Structures ğŸŒ
data = {
    "students": [
        {"name": "John", "marks": [90, 85, 82]},
        {"name": "Emma", "marks": [95, 92, 88]}
    ]
}
print(data["students"][1]["marks"][0])  # Access nested element
```
