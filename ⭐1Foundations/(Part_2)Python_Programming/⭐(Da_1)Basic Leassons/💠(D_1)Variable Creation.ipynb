# ğŸ“ Variables, Basic Operators & Built-in Functions

```python
# Variable assignment
name = "Alice"
age = 25

# Basic operators
sum_age = age + 5       # Addition â•
is_adult = age >= 18    # Comparison âœ…

# Built-in functions
print(f"Name: {name}, Age: {age}")  # Output to console
print(len(name))                     # Length of string
print(type(age))                     # Type of variable
```

```python
# Assign different values to each variable
x, y, z = 1, 2, 3

print(x)  # Output: 1
print(y)  # Output: 2
print(z)  # Output: 3

# Assign the same value to multiple variables
x = y = z = 1

print(x)  # Output: 1
print(y)  # Output: 1
print(z)  # Output: 1
```

# ğŸ—‚ï¸ Python Operators

```python
# 1ï¸âƒ£ Arithmetic Operators â•â–âœ–ï¸â—
a = 10
b = 3
print("Addition:", a + b)      # 13
print("Subtraction:", a - b)   # 7
print("Multiplication:", a * b) # 30
print("Division:", a / b)       # 3.3333
print("Floor Division:", a // b) # 3
print("Modulus:", a % b)        # 1
print("Exponent:", a ** b)      # 1000
```

```python

# 2ï¸âƒ£ Assignment Operators ğŸ“
x = 5
x += 3   # x = x + 3
x -= 2   # x = x - 2
x *= 4   # x = x * 4
x /= 3   # x = x / 3
print("Updated x:", x)
```

```python

# 3ï¸âƒ£ Comparison Operators âš–ï¸
p = 7
q = 5
print("Equal:", p == q)       # False
print("Not equal:", p != q)   # True
print("Greater than:", p > q) # True
print("Less than:", p < q)    # False
print("Greater or equal:", p >= q) # True
print("Less or equal:", p <= q)    # False
```

```python

# 4ï¸âƒ£ Logical Operators ğŸ”—
x = True
y = False
print("AND:", x and y)   # False
print("OR:", x or y)     # True
print("NOT x:", not x)   # False
```

```python

# 5ï¸âƒ£ Membership Operators ğŸ§©
my_list = [1, 2, 3, 4]
print("2 in list:", 2 in my_list)      # True
print("5 not in list:", 5 not in my_list) # True
```

```python

# 6ï¸âƒ£ Identity Operators ğŸ†”
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print("a is b:", a is b)  # False (different objects)
print("a is c:", a is c)  # True
print("a is not b:", a is not b) # True
```

```python

# ğŸ”¹ Quick Tips:
# - Arithmetic â†’ math operations
# - Assignment â†’ update variables
# - Comparison â†’ True/False checks
# - Logical â†’ combine conditions
# - Membership â†’ check presence
# - Identity â†’ check object references
```

# ğŸ—‚ï¸ Python Data Types



```python

# 1ï¸âƒ£ Numeric Types ğŸ”¢
x_int = 10          # int â†’ integer
x_float = 3.14      # float â†’ decimal number
x_complex = 2 + 3j  # complex â†’ real + imaginary
print(type(x_int), type(x_float), type(x_complex))
```

```python

# 2ï¸âƒ£ Text Type ğŸ“
name = "Alice"      # str â†’ string
print(type(name))
print("First character:", name[0])
print("Length:", len(name))
```

```python

# 3ï¸âƒ£ Sequence Types ğŸ“œ
# List â€“ ordered, mutable
fruits = ["apple", "banana", "cherry"]
# Tuple â€“ ordered, immutable
coordinates = (10, 20)
# Range â€“ ordered sequence of numbers
nums = range(5)
print(fruits, coordinates, list(nums))
```

```python

# 4ï¸âƒ£ Set Types ğŸ”¹
# Set â€“ unordered, unique
colors = {"red", "green", "blue"}
# Frozenset â€“ immutable set
frozen_colors = frozenset(["red", "green"])
print(colors, frozen_colors)
```

```python

# 5ï¸âƒ£ Mapping Type ğŸ—ºï¸
# Dictionary â€“ key-value pairs
person = {"name": "Bob", "age": 30}
print(person, person["name"])
```

```python

# 6ï¸âƒ£ Boolean Type âœ…âŒ
is_active = True    # True or False
print(type(is_active))
```

```python

# 7ï¸âƒ£ Binary Types ğŸ’¾
b = b"Hello"        # bytes
byte_array = bytearray([65, 66, 67])
memory_view = memoryview(b"ABC")
print(b, byte_array, memory_view)
```

```python

# ğŸ”¹ Quick Tips:
# - Use int/float for numbers, str for text
# - Use list/tuple for sequences
# - Use set/frozenset for unique items
# - Use dict for structured key-value data
# - Use bool for True/False flags
# - Use bytes/bytearray/memoryview for binary data
```

# ğŸ—‚ï¸ Python Built-in Functions


```python

# 1ï¸âƒ£ Type Conversion Functions ğŸ”„
x = "10"
y = int(x)           # str â†’ int
z = float(x)         # str â†’ float
b = str(25)          # int â†’ str
print(y, z, b, type(y), type(z), type(b))
```

```python

# 2ï¸âƒ£ Mathematical Functions â•â–âœ–ï¸â—
a = abs(-7)          # Absolute value
b = pow(2, 3)        # 2^3 = 8
c = round(3.678, 2)  # Round to 2 decimal places
d = min(5, 10, 2)    # Minimum value
e = max(5, 10, 2)    # Maximum value
print(a, b, c, d, e)
```

```python

# 3ï¸âƒ£ Sequence Functions ğŸ“œ
lst = [1, 2, 3, 4]
tup = (5, 6, 7)
print(len(lst))      # Length of sequence
print(sum(lst))      # Sum of elements
print(sorted(tup))   # Sort sequence
print(reversed(lst)) # Reverse iterator
```

```python

# 4ï¸âƒ£ Utility Functions ğŸ› ï¸
name = "Alice"
print(type(name))    # Type of variable
print(isinstance(name, str))  # Check type
print(dir(name))     # List attributes & methods
```

```python

# 5ï¸âƒ£ Input & Output ğŸ’¬
# user_input = input("Enter your name: ")  # Take input from user
print("Hello,", "World!")  # Print output
```

```python

# 6ï¸âƒ£ Other Useful Built-ins ğŸŒŸ
print(bool(0), bool(5))    # Convert to boolean
print(chr(65), ord('A'))   # char â†” ASCII
print(range(5))             # Generate range
print(help(len))            # Get documentation
```

```python

# ğŸ”¹ Quick Tips:
# - Use int(), float(), str() to convert types
# - Use abs(), pow(), round(), min(), max() for math
# - Use len(), sum(), sorted(), reversed() for sequences
# - Use type(), isinstance(), dir(), help() for inspection
# - Use input() & print() for I/O
```
