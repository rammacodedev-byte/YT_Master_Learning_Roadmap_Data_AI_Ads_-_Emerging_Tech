```python

# üîπ Quick Tips:
# - Inheritance allows code reuse
# - Child class can override parent methods
# - super() accesses parent class methods
# - Python supports multiple inheritance
# - MRO decides which method is called
```

```python

# 1Ô∏è‚É£ Parent (Base) Class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Animal makes a sound")
        
```

```python

# 2Ô∏è‚É£ Child (Derived) Class
class Dog(Animal):
    def speak(self):                  # Method overriding
        print(f"{self.name} says Woof!")
```

```python

# 3Ô∏è‚É£ Create Object of Child Class
dog = Dog("Buddy")
dog.speak()                           # Uses child class method
```

```python

# 4Ô∏è‚É£ Using super() üîó
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)        # Call parent constructor
        self.color = color

    def speak(self):
        print(f"{self.name} says Meow!")

cat = Cat("Kitty", "White")
cat.speak()
```

```python

# 5Ô∏è‚É£ Types of Inheritance üìö

# Single Inheritance
class A:
    pass

class B(A):
    pass

# Multilevel Inheritance
class C(B):
    pass

# Multiple Inheritance
class X:
    def show(self):
        print("Class X")

class Y:
    def show(self):
        print("Class Y")

class Z(X, Y):
    pass

z = Z()
z.show()    # Method Resolution Order (MRO)
```
