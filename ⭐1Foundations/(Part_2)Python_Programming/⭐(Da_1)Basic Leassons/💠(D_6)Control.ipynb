# üóÇÔ∏è Python Control Flow Statements


```python

# üîπ Quick Tips:
# - if ‚Üí check a condition
# - elif ‚Üí multiple conditions
# - else ‚Üí default case
# - Nested if ‚Üí conditions inside conditions
# - pass ‚Üí empty block placeholder
# - match-case ‚Üí cleaner alternative to multiple elif
```

```python
# 1Ô∏è‚É£ if Statement üîç
age = 18
if age >= 18:
    print("You are eligible to vote")
```

```python

# 2Ô∏è‚É£ if‚Äìelse Statement üîÑ
number = 7
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python

# 3Ô∏è‚É£ if‚Äìelif‚Äìelse Statement üß≠
marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Fail")
```

```python

# 4Ô∏è‚É£ Nested if ü™ú
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Invalid user")
```

```python

# 5Ô∏è‚É£ pass Statement üöß
x = 10
if x > 5:
    pass    # Placeholder, no action yet
```

```python

# 6Ô∏è‚É£ match-case (Python 3.10+) üéØ
day = 3
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day")
```
