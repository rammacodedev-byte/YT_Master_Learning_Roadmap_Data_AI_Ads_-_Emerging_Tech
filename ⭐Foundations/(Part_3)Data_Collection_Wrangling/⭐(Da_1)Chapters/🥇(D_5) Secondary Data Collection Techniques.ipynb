```python

# 1Ô∏è‚É£ Definition
definition = """
Secondary Data Collection refers to gathering data from existing sources 
that was collected by someone else for a different purpose. 
It is faster and cheaper but may not be perfectly tailored.
"""
print(definition)
```

```python

# 2Ô∏è‚É£ Techniques Overview
techniques = {
    "Published Reports": "Books, research papers, journals, magazines providing verified data.",
    "Databases": "Government records, company databases, public repositories (e.g., Kaggle, UCI ML repository).",
    "Websites / Online Sources": "Scraping or downloading data from websites, portals, APIs.",
    "Social Media": "Data from Twitter, Instagram, LinkedIn, Facebook for analytics.",
    "Media Sources": "News articles, TV reports, podcasts, blogs providing insights."
}

print("Secondary Data Collection Techniques:")
for method, desc in techniques.items():
    print(f"- {method}: {desc}")
```
