```python

# 1ï¸âƒ£ Definition
definition = """
Data Wrangling is the process of cleaning, transforming, and organizing raw data 
into a structured and usable format for analysis or machine learning.
It typically follows multiple phases to ensure data quality and readiness.
"""
print(definition)
```

```python

# 2ï¸âƒ£ Phases Overview
phases = {
    "Data Discovery": "Understand the dataset: check structure, types, missing values, and anomalies",
    "Data Cleaning": "Handle missing values, duplicates, outliers, and incorrect data",
    "Data Transformation": "Convert data types, normalize or scale features, encode categorical variables",
    "Data Enrichment": "Combine multiple datasets, create new features, augment data",
    "Data Validation": "Ensure consistency, correctness, and integrity of the data",
    "Data Storage / Export": "Store cleaned data in files or databases for further analysis"
}

print("Phases of Data Wrangling:")
for phase, desc in phases.items():
    print(f"- {phase}: {desc}")
```

```python

# 3ï¸âƒ£ Python Example â€“ Simple Wrangling Flow
import pandas as pd
import numpy as np

# Raw Data
data = {
    "Name": ["Alice", "Bob", None, "Emma", "John", "Alice"],
    "Age": [25, 30, 22, None, 28, 25],
    "City": ["Paris", "London", "New York", "paris", None, "Paris"]
}
df = pd.DataFrame(data)
print("\nRaw Data:")
print(df)
```

```python
# Phase 1: Data Cleaning
df["Name"].fillna("Unknown", inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["City"].fillna("Unknown", inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# Phase 2: Data Transformation
df["City"] = df["City"].str.title()        # Standardize text
df["Name"] = df["Name"].str.strip()
```

```python
# Phase 3: Data Validation
assert df["Age"].min() >= 0                # Example validation
```

```python

# Phase 4: Data Ready
print("\nCleaned & Transformed Data:")
print(df)
```
