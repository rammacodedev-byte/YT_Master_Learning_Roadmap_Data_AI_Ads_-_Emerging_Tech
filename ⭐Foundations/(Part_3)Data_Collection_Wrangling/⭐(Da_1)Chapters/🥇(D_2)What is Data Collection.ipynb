```python

# 1Ô∏è‚É£ What is Data Collection?
# Data collection is the process of gathering raw data from different sources
# to analyze, understand, and build predictive models.

definition = """
Data Collection is the process of acquiring accurate, complete, and relevant
data from various sources to use for analysis, reporting, or machine learning.
"""
print(definition)
```

```python

# 2Ô∏è‚É£ Types of Data Sources üèûÔ∏è

# a) Structured Data
structured_data = {
    "CSV / Excel files": "Local datasets with rows and columns",
    "Databases": "SQL, NoSQL, cloud databases"
}

# b) Unstructured Data
unstructured_data = [
    "Text files / PDFs",
    "Images / Videos",
    "Audio / Speech",
    "Social media posts"
]

print("Structured Data Sources:", structured_data)
print("Unstructured Data Sources:", unstructured_data)

# c) Real-time / Streaming Data
streaming_sources = ["IoT sensors", "Web logs", "APIs"]
print("Streaming Data Sources:", streaming_sources)
```

```python

# 3Ô∏è‚É£ Methods of Data Collection üß∞
methods = [
    "Manual data entry",
    "Web scraping / APIs",
    "Sensors / IoT devices",
    "Surveys / Questionnaires",
    "Databases / Logs",
]

for idx, method in enumerate(methods, 1):
    print(f"{idx}. {method}")
```

```python

# 4Ô∏è‚É£ Quick Tips üîπ
# - Ensure data is relevant to the problem
# - Check data quality (accuracy, completeness, consistency)
# - Keep data in accessible formats (CSV, Excel, JSON)
# - Respect privacy & ethics when collecting data
# - Always document the source and method
```
