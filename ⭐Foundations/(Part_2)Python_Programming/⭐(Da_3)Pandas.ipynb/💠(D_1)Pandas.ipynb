```python
# üîπ Quick Tips:
# - DataFrame ‚Üí main Pandas object
# - head(), info(), describe() for EDA
# - iloc ‚Üí index-based, loc ‚Üí label-based
# - groupby() for aggregation
# - read_csv() / to_csv() for files
```

```python
import pandas as pd

# 1Ô∏è‚É£ Create a DataFrame üìä
data = {
    "Name": ["Alice", "Bob", "Emma"],
    "Age": [25, 30, 22],
    "City": ["Paris", "London", "New York"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 2Ô∏è‚É£ Basic Data Exploration üîç
print(df.head())        # First 5 rows
print(df.tail())        # Last 5 rows
print(df.shape)         # Rows, Columns
print(df.info())        # Data info
print(df.describe())    # Statistical summary
```

```python
# 3Ô∏è‚É£ Select Data üéØ
print(df["Name"])               # Select column
print(df[["Name", "City"]])     # Multiple columns
print(df.iloc[0])               # Select by index
print(df.loc[0, "Age"])         # Select by label
```

```python
# 4Ô∏è‚É£ Filtering Data üßπ
print(df[df["Age"] > 25])
```

```python
# 5Ô∏è‚É£ Add & Modify Columns ‚ûï
df["Salary"] = [50000, 60000, 55000]
df["Age"] = df["Age"] + 1
print(df)
```

```python
# 6Ô∏è‚É£ Handle Missing Values üöß
df.loc[1, "City"] = None
print(df.isnull())
print(df.fillna("Unknown"))
```

```python
# 7Ô∏è‚É£ Sort & Group üìà
print(df.sort_values("Age"))
print(df.groupby("City")["Salary"].mean())
```
