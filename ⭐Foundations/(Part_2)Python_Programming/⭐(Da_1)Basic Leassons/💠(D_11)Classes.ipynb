```python

# üîπ Quick Tips:
# - Class ‚Üí blueprint for objects
# - Object ‚Üí instance of class
# - __init__ ‚Üí constructor method
# - Inheritance ‚Üí reuse code
# - Encapsulation ‚Üí hide private data
# - Polymorphism ‚Üí same method, different behavior
```

```python

# 1Ô∏è‚É£ Define a Class üèóÔ∏è
class Person:
    # Constructor (__init__) to initialize object
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Method
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
```

```python

# 2Ô∏è‚É£ Create Objects üß©
p1 = Person("Alice", 25)
p2 = Person("Bob", 30)
```

```python

# 3Ô∏è‚É£ Access Object Properties & Methods
print(p1.name)
print(p2.age)
p1.greet()
p2.greet()
```

```python

# 4Ô∏è‚É£ Inheritance üå≥
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)   # Call parent constructor
        self.student_id = student_id

    def show_id(self):
        print(f"Student ID: {self.student_id}")

s1 = Student("Emma", 20, "S123")
s1.greet()        # Inherited method
s1.show_id()      # Child method
```

```python

# 5Ô∏è‚É£ Encapsulation üîí
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())   # 1500
# print(account.__balance)     # ‚ùå Error: private variable
```

```python

# 6Ô∏è‚É£ Polymorphism üîÑ
class Dog:
    def speak(self):
        print("Woof!")

class Cat:
    def speak(self):
        print("Meow!")

def animal_sound(animal):
    animal.speak()   # Same method name, different implementation

dog = Dog()
cat = Cat()
animal_sound(dog)
animal_sound(cat)
```
