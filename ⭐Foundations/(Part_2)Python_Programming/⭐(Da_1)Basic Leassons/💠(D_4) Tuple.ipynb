# üóÇÔ∏è Python Tuple ‚Äì Ordered & Immutable üîí

```python

# üîπ Quick Tips:
# - Tuples are ordered and immutable
# - Faster and safer than lists for fixed data
# - Use tuples for constants and records
# - Single-element tuple needs a comma (value,)
```

```python


# 1Ô∏è‚É£ Create a Tuple
numbers = (10, 20, 30)
single_item = (5,)          # Single-element tuple (comma is required)
print(numbers)
print(single_item)
```

```python

# 2Ô∏è‚É£ Access Tuple Elements
print(numbers[0])           # First element
print(numbers[-1])          # Last element
print(numbers[1:3])         # Slicing
```

```python

# 3Ô∏è‚É£ Tuple is Immutable
# numbers[0] = 100           # ‚ùå Error: tuples cannot be modified
```

```python

# 4Ô∏è‚É£ Tuple Operations
print(len(numbers))         # Length of tuple
print(20 in numbers)        # Membership check
print(numbers.count(20))    # Count occurrences
print(numbers.index(30))    # Find index of element
```

```python

# 5Ô∏è‚É£ Tuple Packing & Unpacking üéÅ
data = ("Alice", 25, "Engineer")
name, age, job = data       # Unpacking
print(name, age, job)
```

```python

# 6Ô∏è‚É£ Nested Tuple
nested = ((1, 2), (3, 4), (5, 6))
print(nested[1][0])         # Access nested element
```
