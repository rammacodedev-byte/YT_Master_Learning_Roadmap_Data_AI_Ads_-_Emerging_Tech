# üóÇÔ∏è Python Functions

```python
# üîπ Quick Tips:
# - def ‚Üí define a function
# - return ‚Üí send value back
# - *args ‚Üí multiple positional arguments
# - **kwargs ‚Üí multiple keyword arguments
# - lambda ‚Üí short one-line function
# - Docstrings ‚Üí explain function purpose
```

```python


# 1Ô∏è‚É£ Define a Function üõ†Ô∏è
def greet():
    print("Hello, Welcome to Python!")

greet()  # Function call
```

```python

# 2Ô∏è‚É£ Function with Parameters üì•
def add(a, b):
    return a + b

result = add(5, 3)
print("Sum:", result)
```

```python

# 3Ô∏è‚É£ Default Parameters ‚öôÔ∏è
def greet_user(name="Guest"):
    print("Hello,", name)

greet_user()
greet_user("Alice")
```

```python

# 4Ô∏è‚É£ Keyword Arguments üß©
def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info(age=20, name="Bob")
```

```python

# 5Ô∏è‚É£ *args (Multiple Arguments) üì¶
def total_sum(*numbers):
    return sum(numbers)

print(total_sum(1, 2, 3, 4))
```

```python

# 6Ô∏è‚É£ **kwargs (Keyword Arguments) üóÇÔ∏è
def profile(**details):
    for key, value in details.items():
        print(key, ":", value)

profile(name="Emma", role="Developer", country="USA")
```

```python

# 7Ô∏è‚É£ Lambda Function ‚ö°
square = lambda x: x * x
print(square(5))
```

```python

# 8Ô∏è‚É£ Docstring üìù
def multiply(a, b):
    """Returns the product of two numbers"""
    return a * b

print(multiply(4, 6))
print(multiply.__doc__)


```
