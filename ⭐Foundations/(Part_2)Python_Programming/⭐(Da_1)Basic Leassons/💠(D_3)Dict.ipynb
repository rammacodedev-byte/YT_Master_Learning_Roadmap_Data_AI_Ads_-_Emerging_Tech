```python

# üîπ Quick Tips:
# - Use dictionaries to store related data as key-value pairs
# - Keys must be immutable (str, int, tuple, etc.)
# - Values can be any data type, including lists, sets, or other dictionaries
# - Use get() to avoid errors when key is missing
# - Useful for structured & fast-access data
```

```python
# üóÇÔ∏è Python Dictionary ‚Äì Key-Value Pairs üóùÔ∏è

# 1Ô∏è‚É£ Create a Dictionary
person = {
    "name": "Alice",
    "age": 25,
    "city": "Paris"
}
print(person)
```

```python

# 2Ô∏è‚É£ Access Values
print(person["name"])       # Access by key
print(person.get("age"))    # Safer access (returns None if key not found)
```

```python

# 3Ô∏è‚É£ Add or Update Values
person["job"] = "Engineer"  # Add new key-value
person["age"] = 26          # Update existing value
print(person)
```

```python

# 4Ô∏è‚É£ Remove Items
person.pop("city")          # Remove by key
print(person)
person.popitem()            # Remove last inserted item
print(person)
# del person["name"]        # Remove specific key
# person.clear()            # Remove all items
```

```python

# 5Ô∏è‚É£ Dictionary Keys, Values, Items
print(person.keys())        # All keys
print(person.values())      # All values
print(person.items())       # All key-value pairs
```

```python

# 6Ô∏è‚É£ Nested Dictionary
students = {
    "John": {"age": 18, "marks": [90, 85]},
    "Emma": {"age": 19, "marks": [95, 92]}
}
print(students["Emma"]["marks"][0])  # Access nested data
```
