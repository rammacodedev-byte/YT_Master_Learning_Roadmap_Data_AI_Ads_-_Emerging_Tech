# üóÇÔ∏è Matplotlib ‚Äì Data Visualization Library üìä


```python

# üîπ Quick Tips:
# - plot() ‚Üí line graph
# - bar() ‚Üí bar chart
# - scatter() ‚Üí relationship between variables
# - hist() ‚Üí data distribution
# - pie() ‚Üí proportion visualization
# - Always use title(), xlabel(), ylabel()
```

```python

import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Line Plot üìà
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

# 2Ô∏è‚É£ Bar Chart üìä
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python

# 3Ô∏è‚É£ Scatter Plot üîµ
x = [1, 2, 3, 4, 5]
y = [30, 25, 40, 35, 50]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

# 4Ô∏è‚É£ Histogram üìâ
data = [10, 20, 20, 30, 30, 30, 40, 50]

plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python

# 5Ô∏è‚É£ Pie Chart ü•ß
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```
