# üóÇÔ∏è Plotly Express ‚Äì Interactive Visualization üìä


```python

# üîπ Quick Tips:
# - px ‚Üí high-level, easy Plotly interface
# - fig.show() ‚Üí display interactive chart
# - Hover, zoom, pan are enabled by default
# - Great for dashboards & EDA
```

```python
import plotly.express as px

# Sample DataFrame
df = px.data.iris()   # Built-in dataset

# 1Ô∏è‚É£ Scatter Plot üîµ
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="Scatter Plot with Plotly Express"
)
fig.show()
```

```python

# 2Ô∏è‚É£ Line Plot üìà
gap = px.data.gapminder().query("country=='India'")
fig = px.line(
    gap,
    x="year",
    y="lifeExp",
    title="Life Expectancy Over Time"
)
fig.show()
```

```python

# 3Ô∏è‚É£ Bar Chart üìä
fig = px.bar(
    df,
    x="species",
    y="petal_length",
    title="Bar Chart Example"
)
fig.show()
```

```python

# 4Ô∏è‚É£ Histogram üìâ
fig = px.histogram(
    df,
    x="sepal_length",
    nbins=20,
    title="Histogram Example"
)
fig.show()
```

```python

# 5Ô∏è‚É£ Box Plot üì¶
fig = px.box(
    df,
    x="species",
    y="sepal_length",
    title="Box Plot Example"
)
fig.show()
```

```python

# 6Ô∏è‚É£ Pie Chart ü•ß
tips = px.data.tips()
fig = px.pie(
    tips,
    names="day",
    values="total_bill",
    title="Pie Chart Example"
)
fig.show()
```
